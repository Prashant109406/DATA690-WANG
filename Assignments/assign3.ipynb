Read (with naked eyes, not computer) the dictionary and find out the definitions of the following 12 columns and copy the text next to the column name
- UNITID: The number or name of a specific unit            
- INSTNM: Institution name     
- CITY:  Name of the city where the institution is.                    
- ZIP:   Zip Code of the place
- ADM_RATE:    Admission rate is the rate at which applicants are accepted      
- UGDS:     Enrollment of undergraduate certificate/degree-seeking students        
- TUITIONFEE_IN: in-state tuition fee      
- MN_EARN_WNE_P10:Mean earnings of students working and not enrolled 10 years after entry

```python
# reimport the data with above 12 columns (ignore all other variables)
file=pd.read_csv(r"C:\Users\Prashant Mourya\Downloads\CollegeScorecard_Raw_Data\MERGED2017_18_PP.csv",low_memory=False,usecols=["UNITID","INSTNM","CITY","ZIP","ADM_RATE","UGDS","TUITIONFEE_IN","MN_EARN_WNE_P10"])
#file=file.fillna(0)
file
```

```python
# Find out the 10 colleges that are hardest to get in (top 10)
top_ten=file.nsmallest(10, 'ADM_RATE')[["INSTNM","ADM_RATE","TUITIONFEE_IN","MN_EARN_WNE_P10"]]
top_ten
```

```python
# 10. Find out the average in-state tuition of top 10

top_fee=top_ten.fillna(0).nlargest(10, 'TUITIONFEE_IN')["TUITIONFEE_IN"]


print("the average in-state tuition of top 10 is",ny.average(top_fee))
```

```python
# Find out the average earnings of graduates from the top 10
top_earning=top_ten.nlargest(10, 'MN_EARN_WNE_P10')["MN_EARN_WNE_P10"]
top_earning
print("average earnings of graduates from the top 10 is",ny.average(top_earning))
```

```python
# Find out the 10 colleges that are easiest to get in (bottom 10)
bottom_ten=file.nlargest(10, 'ADM_RATE')[["INSTNM","ADM_RATE","TUITIONFEE_IN","MN_EARN_WNE_P10"]]
bottom_ten
```

```python
# Find out the average in-state tuition of the bottom 10
top_fee=bottom_ten.fillna(0).nlargest(10, 'TUITIONFEE_IN')["TUITIONFEE_IN"]

print("the average in-state tuition of the bottom 10 is",ny.average(top_fee))
```

```python
# Find out the average earnings of graduates from the bottom 10
top_earning=bottom_ten.nlargest(10, 'MN_EARN_WNE_P10')["MN_EARN_WNE_P10"]
top_earning
print("the average earnings of graduates from the bottom 10 is",ny.average(top_earning))
```
